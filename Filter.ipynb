# FILTER

``# SYNTAX``

>filter(função, iterável)

Filtra os dados retornados de acordo com a sua necessidade

```python
produtos = [
    {'preço': 55.0,   'peso': 1.30},
    {'preço': 46.5,   'peso': 2.41},
    {'preço': 22.0,   'peso': 3.20},
    {'preço': 15.5,   'peso': 1.44},
    {'preço': 43.1,   'peso': 2.34},
    {'preço': 46.5,   'peso': 2.41},
    {'preço': 14.2,   'peso': 1.00},
    {'preço': 17.5,   'peso': 1.24},
    {'preço': 242.0,  'peso': 4.14},
    {'preço': 1200.5, 'peso': 16.10}
]

produtos_filter = list( # Convertendo para lista
    map( # Aplicando a função Map para percorrer o iterável
        lambda p: p['preço'], # Passando nossa função lambda como parâmetro para exibir somente o preço.
        filter(lambda p: p['preço'] > 20, produtos ) 
    )
)
print(produtos_filter)
```

Outra maneira de escrever o código final acima

```python
produtos_filter = list(map(lambda p: p['preço'],
                           filter(lambda p: p['preço'] > 20, produtos )))
```

Usando apenas o filter, sem usar map para exibir somente os valores

```python
produtos_filter = list(filter(lambda p: p['preço'] > 20, produtos ))
print(produtos_filter)
```

### No Python Moderno é mais comum usar Compreensão de Listas ao invés de map e filter!
