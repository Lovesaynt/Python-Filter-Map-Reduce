# REDUCE

Reduz uma sequência de valores a um único resultado, aplicando uma função acumuladora

``SYNTAX``

import reduce

>reduce(função, iterável, initial=0)

```python
from functools import reduce

produtos = [
    {'preço': 55.0,   'peso': 1.30},
    {'preço': 46.5,   'peso': 2.41},
    {'preço': 22.0,   'peso': 3.20},
    {'preço': 15.5,   'peso': 1.44},
    {'preço': 43.1,   'peso': 2.34},
    {'preço': 46.5,   'peso': 2.41},
    {'preço': 14.2,   'peso': 1.00},
    {'preço': 17.5,   'peso': 1.24},
    {'preço': 242.0,  'peso': 4.14},
    {'preço': 1200.5, 'peso': 16.10}
]

preço_total = reduce(lambda ac, p: ac +p['preço'], produtos, 0)

print(preço_total)
```
