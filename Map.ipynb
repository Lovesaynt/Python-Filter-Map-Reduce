# MAP

``# SYNTAX``

>map(função, iterável)

Aplica uma função a um iterável.

    Retorna um map object num espaço na memória
    Deve ser convertido para lista

```python
preços = [100, 200, 300, 400, 500] # Iterável

def adicionar_imposto(preço): # Função de Imposto
    return preço * 1.1
```

    Exemplo com FOR

```python
p_imposto = []
for p in preços: # Aplicando a função de Imposto com FOR
    p_imposto.append(adicionar_imposto(p)) # Adicionando a Lista

print(p_imposto)
```

    Usando o Map

```python
p_imposto2 = list(map(adicionar_imposto, preços)) # Mesmo resultado acima, porém mais compacto com o uso do Map.
print(p_imposto2)
```

### No Python Moderno é mais comum usar Compreensão de Listas ao invés de map e filter!
